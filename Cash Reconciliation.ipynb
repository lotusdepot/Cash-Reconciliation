```python
# Subtype 	 Type 	 Date 	 Date 	 Sequence 	 Cashier Name 	 Custom Data 	 USD Amount 

# END_OF_SHIFT = SUM by DATE( USD Amount , Type = "END_OF_SHIFT")+ SUM by DATE (USD AMOUNT, Type = "DEPOSIT", subtype is "" (blank))
# Tills Pulled = SUM by DATE( all BLANK "" subtype, type = DISPENSE) - SUBTRACT (DISPENSE with no custom data (Need User Input))
# Total Cash Sales = End of Shift + Tills
```

```python
#imports
import pandas as pd
import numpy as np
```

```python
cash_recon = pd.read_excel("May 22 Raw Cash Treez Data.xlsx",[0,1])
raw_data = cash_recon[0][["Subtype", "Type", "Datetime", "Sequence", "Cashier Name", "Custom Data", "USD Amount"]].copy()
raw_data["Date"] = pd.to_datetime(raw_data["Datetime"])
cashierlist = ["Ballesteros, Michael","MCGEE, MIKE","Hernandez, Melissa","SERENO, CLAUDIA", "SANDERS, ERNIE", "TUCCI, NICK"]

recon = pd.DataFrame(pd.Series(raw_data["Date"].dt.date.unique(),name="Date")).set_index("Date")

# raw_data.fillna(0)

recon["End of Shift"] = 0
recon["Tills Pulled"] = 0
for index, row in raw_data.iterrows():
#     print(row["Cashier Name"])
    prev = recon.loc[recon.index[row["Date"].day-1], "End of Shift"]
    dollars = [0 if np.isnan(row["USD Amount"]) else float(row["USD Amount"])][0]
#     print(dollars)

    #End of Shift calculation
    if row["Type"]=="END_OF_SHIFT":
        recon.loc[recon.index[row["Date"].day-1], "End of Shift"] = float(prev) + dollars
    if row["Type"]=="DEPOSIT" and not isinstance(row["Subtype"], str):
        recon.loc[recon.index[row["Date"].day-1], "End of Shift"] = float(prev) + dollars
    
    prev = recon.loc[recon.index[row["Date"].day-1], "Tills Pulled"]
    USDList = [-400, -174, -225]
    #Tills Pulled calculation
    if row["Type"]=="DISPENSE" and not isinstance(row["Subtype"], str):
        if not(row["Custom Data"]=="{}" and row["Cashier Name"] in cashierlist):
            recon.loc[recon.index[row["Date"].day-1], "Tills Pulled"] = float(prev) + dollars
        else:
            if row["Cashier Name"] == "SERENO, CLAUDIA" and row["USD Amount"] in USDList:
                recon.loc[recon.index[row["Date"].day-1], "Tills Pulled"] = float(prev) + dollars
display(recon)
# raw_data.to_csv("May 2022 CashReconScript Export.csv")
```
